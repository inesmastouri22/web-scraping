```python
table_data = []
for j in ['DE', 'AT', 'CH']:
    url = f'https://ttlist.net/tiktok/top?countries={j}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    rows = soup.find_all('tr')  # Assuming you're getting the rows from the table

    for row in rows:
        td_elements = row.find_all('td')
        if len(td_elements) >= 5:
            id_ = row.get('id', 'N/A')
            position = td_elements[0].text.strip()
            profile_link_element = row.find('a', href=True)
            profile_link = profile_link_element['href'] if profile_link_element else 'N/A'
            likes = td_elements[2].text.strip()
            followers = td_elements[3].text.strip()
            country = td_elements[4].text.strip()

            subject_category_element = td_elements[5].find('a') if len(td_elements) > 5 else None
            subject_category = subject_category_element.text.strip() if subject_category_element else 'N/A'

            table_data.append({
                'id': id_,
                'position': position,
                'profile_link': profile_link,
                'likes': likes,
                'followers': followers,
                'country': country,
                'subject_category': subject_category
            })

# Print the data
for data in table_data:
    print(data)
```
