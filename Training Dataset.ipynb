```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd

# Path to the ChromeDriver executable
chromedriver_path = 'C:/Users/DELL/Downloads/chromedriver-win64/chromedriver-win64/chromedriver.exe'  # Update this to the actual path

# Set up Chrome options for headless mode
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--window-size=1920x1080")
chrome_options.add_argument('--disable-blink-features=AutomationControlled')

# Set up the Selenium driver service
chrome_service = Service(chromedriver_path)

# Set up the Selenium driver
driver = webdriver.Chrome(service=chrome_service, options=chrome_options)

# Base URL for the pages to scrape
base_url = 'https://www.udacity.com/catalog/all/any-price/cloud%20computing,cybersecurity,business,product%20management,Executive%20Leadership,programming%20and%20development,artificial%20intelligence,autonomous%20systems,data%20science,Career%20resources/any-skill/any-difficulty/any-duration/any-type/highest-rated/page-'

# Number of pages to scrape
num_pages = 24

# List to store course data
course_data = []

# Loop through each page number
for i in range(1, num_pages + 1):
    url = f'{base_url}{i}'
    print(f'Scraping URL: {url}')

    # Open the webpage
    driver.get(url)

    # Explicitly wait for the elements to be present
    try:
        WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'a.chakra-heading.css-1rsglaw'))
        )

        # Get the page source and parse with BeautifulSoup
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')

        # Find all course elements
        course_elements = soup.find_all('a', class_='chakra-heading css-1rsglaw')

        # Extract the required details from each course element
        for course_element in course_elements:
            text = course_element.text.strip()
            href = 'https://www.udacity.com' + course_element.get('href', 'No href attribute')

            # Navigate to the course details page to extract additional information
            driver.get(href)
            WebDriverWait(driver, 30).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'p.chakra-text.css-5ucqax'))
            )
            course_page_source = driver.page_source
            course_soup = BeautifulSoup(course_page_source, 'html.parser')

            # Extract difficulty level
            difficulty_level_element = course_soup.find('p', class_='chakra-text css-5ucqax')
            difficulty_level = difficulty_level_element.text.strip() if difficulty_level_element else 'N/A'

            # Extract course duration
            duration_element = course_soup.find('p', class_='chakra-text css-1osqk4n')
            duration = duration_element.text.strip() if duration_element else 'N/A'

            # Extract rating
            rating_element = course_soup.find('div', {'aria-label': True})
            rating = rating_element['aria-label'].strip() if rating_element else 'N/A'

            # Append the extracted data to the list
            course_data.append((text, href, difficulty_level, duration, rating))

    except Exception as e:
        print(f"Error scraping {url}: {e}")

# Close the Selenium driver
driver.quit()

# Create a DataFrame and display it
df = pd.DataFrame(course_data, columns=['Course Name', 'Href', 'Difficulty Level', 'Duration', 'Rating'])
print(df)

# Save the DataFrame to a CSV file
df.to_csv('C:/Users/DELL/Desktop/Stage/Results/Training Dataset1.csv', index=False)
```
